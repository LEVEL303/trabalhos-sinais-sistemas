### 1. Considere o sinal de tempo contı́nuo definido pela expressão 𝑥(𝑡) = (𝑡² − 1) [𝑢(𝑡 − 1) − 𝑢(𝑡 − 2)].

#### (a) Plote o sinal 𝑥(𝑡) para 0 ≤ 𝑡 ≤ 3.

```python
import numpy as np
import matplotlib.pyplot as plt

t_inicio = 0
t_fim = 3
npoints = 1000
t = np.linspace(t_inicio, t_fim, npoints)

x = (t**2 - 1) * ((t>=1) & (t<2))

plt.figure(figsize=(4,3))
plt.plot(t, x)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.xlim(t_inicio, t_fim)
plt.grid()
plt.show()
```

#### (b) Calcule, de forma analı́tica, a energia do sinal 𝑥(𝑡).

A energia é dada por:

$$
Ex = \int_{-\infty}^{\infty} |x(t)|^2 dt
$$

Como o sinal é diferente de zero apenas em $1 \le t \le 2$:

$$
Ex = \int_{1}^{2} (t^2 - 1)^2 dt
$$

Expandindo:

$$
(t^2 - 1)^2 = t^4 - 2t^2 + 1
$$

Substituindo na integral:

$$
Ex = \int_{1}^{2} t^4 - 2t^2 + 1 dt
$$

Calculando integral:

$$
Ex = \int_{1}^{2} t^4 - 2t^2 + 1 dt = \left[ \frac{1}{5} \cdot t^5 - \frac{2}{3} \cdot t^3 + t \right]_{1}^{2}
$$

Aplicando os limites de integração:

$$
Ex = \left(\frac{1}{5} \cdot 2^5 - \frac{2}{3} \cdot 2^3 + 2 \right) - \left(\frac{1}{5} \cdot 1^5 - \frac{2}{3} \cdot 1^3 + 1 \right)
= \frac{46}{15} - \frac{8}{15} = \frac{38}{15}
$$

Resultado:

$$
Ex = \frac{38}{15} = 2.533333...
$$

#### (c) Determine, de forma numérica, a energia do sinal 𝑥(𝑡). Compare com o resultado do item anterior.

```python
dt = t[1] - t[0]
Ex = np.sum(x**2) * dt
erro = np.abs(Ex - 38/15)

print("Energia (computacional):", Ex)
print("Energia (analítica):", 38/15)
print("Erro:", erro)
```

$$
x(t) = 2 \cdot [u(t + 1,5) - u(t)] + 2e^{-t/2} \cdot [u(t) - u(t - 3)]
$$

#### (b) Plote o sinal 𝑥(𝑡).

```python
t_inicio = -6
t_fim = 7
npoints = 1000
t = np.linspace(t_inicio, t_fim, npoints)

def meusinal(t):
    return 2 * ((t>=-1.5) & (t<0)) + 2 * np.exp(-t/2) * ((t>=0) & (t<3))

x = meusinal(t)

plt.figure(figsize=(6, 3))
plt.plot(t, x)
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.xlim(t_inicio, t_fim)
plt.grid()
plt.show()
```

(v) 𝑥(𝑡/2)

```python
# x(-t) operação de reflexão

x = meusinal(-t)

plt.figure(figsize=(11, 5))

plt.subplot(2, 3, 1)
plt.plot(t, x)
plt.title('$x(-t)$')
plt.xlabel('$t$')
plt.xlim(t_inicio, t_fim)
plt.grid()

# x(t+4) operação de deslocamento à esquerda em 4 unidades

x = meusinal(t+4)

plt.subplot(2, 3, 2)
plt.plot(t, x)
plt.title('$x(t+4)$')
plt.xlabel('$t$')
plt.xlim(t_inicio, t_fim)
plt.grid()

# x(x-2) operação de deslocamento à direita em 2 unidades

x = meusinal(t-2)

plt.subplot(2, 3, 3)
plt.plot(t, x)
plt.title('$x(t-2)$')
plt.xlabel('$t$')
plt.xlim(t_inicio, t_fim)
plt.grid()

# x(2t) operação de compressão por um fator 2

x = meusinal(2*t)

plt.subplot(2, 3, 4)
plt.plot(t, x)
plt.title('$x(2t)$')
plt.xlabel('$t$')
plt.xlim(t_inicio, t_fim)
plt.grid()

# x(t/2) operação de expansão por um fator 2

x = meusinal(t/2)

plt.subplot(2, 3, 5)
plt.plot(t, x)
plt.title('$x(t/2)$')
plt.xlabel('$t$')
plt.xlim(t_inicio, t_fim)
plt.grid()

plt.tight_layout()
plt.show()
```

##### (iii) x(t − 2)
Deslocamento do sinal em 2 unidades para a direita (atraso no tempo).

##### (iv) x(2t)
Compressão no tempo por um fator 2 (o sinal fica mais “rápido”).

##### (v) x(t/2)
Expansão no tempo por um fator 2 (o sinal fica mais “lento”).

### 4. Um determinado sistema linear e invariante no tempo possui uma resposta ao impulso ℎ(𝑡) dada por 
### $ℎ(𝑡) = 𝑒^{−2𝑡} 𝑢(𝑡)$.

#### (a) Plote a resposta ao impulso do sistema para 0 ≤ 𝑡 ≤ 3.

```python
t_inicio = 0
t_fim = 3
npoints = 1000
t = np.linspace(t_inicio, t_fim, npoints)

h = np.exp(-2*t) * (t>=0)

plt.figure(figsize=(6,3))
plt.plot(t, h)
plt.xlabel('$t$')
plt.ylabel('$h(t)$')
plt.xlim(t_inicio, t_fim)
plt.grid()
plt.show()
```

#### (b) O sistema é estável? O sistema é causal? Justifique suas respostas.

O sistema é causal, pois h(t) = 0 para t < 0 devido ao termo u(t).

O sistema é estável, pois a integral de |h(t)| é finita:
$$
\int_{-\infty}^{\infty} |h(t)| dt = \int_0^\infty e^{-2t} dt < \infty
$$

#### (c) Considere que o seguinte sinal:
#### 𝑥(𝑡) = 𝑢(𝑡) − 𝑢(𝑡 − 2)
#### é aplicado na entrada do sistema. Determine, de forma analı́tica, a expressão da saı́da 𝑦(𝑡) para todo 𝑡 utilizando o método gráfico da convolução.

A saída é dada por:

$$
y(t) = x(t) * h(t)
$$

com

$$
x(t) = u(t) - u(t - 2), \quad h(t) = e^{-2t}u(t)
$$

Logo,

$$
y(t) = \int_{-\infty}^{\infty} x(\tau)h(t-\tau)\,d\tau
$$

Como $x(\tau) \neq 0$ apenas em $0 \le \tau \le 2$, temos:

- Para $t < 0$:
$$
y(t) = 0
$$

- Para $0 \le t \le 2$:
$$
y(t) = \int_0^t e^{-2(t-\tau)} d\tau = \frac{1 - e^{-2t}}{2}
$$

- Para $t > 2$:
$$
y(t) = \int_0^2 e^{-2(t-\tau)} d\tau = \frac{e^{-2(t-2)} - e^{-2t}}{2}
$$

Portanto,

$$
y(t) =
\begin{cases}
0, & t < 0 \\
\frac{1 - e^{-2t}}{2}, & 0 \le t \le 2 \\
\frac{e^{-2(t-2)} - e^{-2t}}{2}, & t > 2
\end{cases}
$$

#### (d) Resolva o item (c) de forma numérica. Compare o resultado numérico com o resultado analı́tico, obtido no item (c). Uma forma de comparar consiste em plotar, em um mesmo gráfico, o sinal da saı́da obtido da forma analı́tica e o sinal de saı́da da forma numérica.

```python
t_inicio = -2
t_fim = 4
npoints = 1000
t = np.linspace(t_inicio, t_fim, npoints)
dt = t[1] - t[0]

# forma numérica
h = np.exp(-2*t) * (t>=0)
x = 1 * ((t>=0) & (t<=2))
y = dt*np.convolve(x, h)

t_inicio_y = 2 * t_inicio
t_fim_y = 2 * t_fim
t_y = np.linspace(t_inicio_y, t_fim_y, len(y))

# forma analítica
y2 = (1 - np.exp(-2*t)) / 2 * ((t>=0) & (t<=2)) + (np.exp(-2*(t-2)) - np.exp(-2*t)) / 2 * (t>2)

plt.figure(figsize=(6,3))
plt.plot(t_y, y, label='Numérico')
plt.plot(t, y2, '--', label='Analítico')
plt.xlabel('$t$')
plt.title('$y(t)$')
plt.xlim(t_inicio_y, t_fim_y)
plt.legend()
plt.grid()
plt.show()
```

#### (e) Repetir os itens (a)-(d) quando a resposta ao impulso do sistema é igual à: $ h(t) = e^t u(t) $

```python
t_inicio = 0
t_fim = 3
npoints = 1000
t = np.linspace(t_inicio, t_fim, npoints)

h = np.exp(t) * (t>=0)

plt.figure(figsize=(6, 3))
plt.plot(t, h)
plt.xlabel('$t$')
plt.ylabel('$h(t)$')
plt.xlim(t_inicio, t_fim)
plt.grid()
plt.show()
```

O sistema é causal, pois h(t) = 0 para t < 0 devido ao termo u(t).

O sistema é instável, pois a integral de |h(t)| é infinita:
$$
\int_{-\infty}^{\infty} |h(t)| dt = \int_0^\infty e^t dt = \infty
$$

A saída é dada por:

$$
y(t) = x(t) * h(t)
$$

com

$$
x(t) = u(t) - u(t - 2), \quad h(t) = e^tu(t)
$$

Logo,

$$
y(t) = \int_{-\infty}^{\infty} x(\tau)h(t-\tau)\,d\tau
$$

Como $x(\tau) \neq 0$ apenas em $0 \le \tau \le 2$, temos:

- Para $t < 0$:
$$
y(t) = 0
$$

- Para $0 \le t \le 2$:
$$
y(t) = \int_0^t e^{(t-\tau)} d\tau = e^t - 1
$$

- Para $t > 2$:
$$
y(t) = \int_0^2 e^{(t-\tau)} d\tau = e^t - e^{t-2}
$$

Portanto,

$$
y(t) =
\begin{cases}
0, & t < 0 \\
e^t - 1, & 0 \le t \le 2 \\
e^t - e^{t-2}, & t > 2
\end{cases}
$$

### 5. Um determinado circuito elétrico com tensão de entrada $v_i(t)$ e tensão de saı́da $v_o(t)$ é modelado através da seguinte equação diferencial ordinária:
$$
\frac{d^2v_o(t)}{dt^2} + 7\frac{dv_o(t)}{dt} + 10v_o(t) = \frac{dv_i(t)}{dt} + 6v_i(t)
$$
### As condições iniciais para esse circuito são $v_o(0) = 6V$ e $\frac{dv_o(0)}{dt} = -4V/s$ e a entrada é $x(t) = 6e^{-3t}u(t)$.

#### (a) Utilizando o método clássico, determine de forma analı́tica a saı́da do sistema $y(t)$ para todo $t$.

Utilizando o método clássico, a saída do sistema $y(t)$ é dada por:
$$
y(t) = y_n(t) + y_f(t)
$$

Calculando a reposta natural:
$$
\ddot{v}_o(t) + 7 \dot{v}_o(t) + 10 v_o(t) \implies \lambda^2 + 7\lambda + 10 = 0
\implies
\begin{cases}
\lambda_1 = -2 \\
\lambda_2 = -5
\end{cases}
$$
$$
y_n(t) = C_1e^{-2t} + C_2e^{-5t}
$$

Calculando a resposta forçada:
$$
x(t) = 6e^{-3t}u(t) \implies y_f(t) = \beta e^{-3t}
$$
$$
\ddot{y}_f(t) + 7\dot{y}_f(t) + 10y_f(t) = \dot{x}(t) + 6x(t)
$$
$$
9\beta e^{-3t} - 21\beta e^{-3t} + 10\beta e^{-3t} = -18e^{-3t} + 36e^{-3t} 
$$
$$
-2\beta e^{-3t} = 18e^{-3t}
$$
$$
\beta = \frac{18e^{-3t}}{-2e^{-3t}} = -9
$$
$$
y_f(t) = -9e^{-3t}
$$

Substituindo as condições iniciais:
$$
y(t) = C_1e^{-2t} + C_2e^{-5t} - 9e^{-3t}
$$
$$
y(0) = 6 = C_1e^{-2 \cdot 0} + C_2e^{-5 \cdot 0} - 9e^{-3 \cdot 0} \implies C_1 + C_2 = 15
$$

$$
\dot{y}(0) = -4 = -2C_1e^{-2 \cdot 0} - 5C_2e^{-5 \cdot 0} + 27e^{-3 \cdot 0} \implies 2C_1 + 5C_2 = 31
$$
$$
\begin{cases}
C_1 + C_2 = 15 \\
2C_1 + 5C_2 = 31
\end{cases}
\quad C_1 = \frac{44}{3} \quad C_2 = \frac{1}{3}
$$

Resposta total:
$$
y(t) = \left(\frac{44}{3} e^{-2t} + \frac{1}{3} e^{-5t} - 9e^{-3t} \right) u(t)
$$

#### (b) Plote o sinal de saı́da para 0 ≤ 𝑡 ≤ 2.

```python
t_inicio = 0
t_fim = 2
npoints = 1000
t = np.linspace(t_inicio, t_fim, npoints)

y = (44/3 * np.exp(-2*t) + 1/3 * np.exp(-5*t) - 9 * np.exp(-3*t)) * (t>=0)

plt.figure(figsize=(6, 3))
plt.plot(t, y)
plt.xlabel('$t$')
plt.title('$y(t)$')
plt.xlim(t_inicio, t_fim)
plt.grid()
plt.show()
```

#### (c) Utilizando computação simbólica, determine a expressão de saı́da para esse circuito quando a entrada é a mesma da apresentada no item (a). Verifique e compare seu resultado com àquele obtido no item (a).

```python
import sympy as sp

t = sp.symbols('t')
x = sp.Function('x')
y = sp.Function('y')

x = 6*sp.exp(-3*t)

eq = sp.diff(y(t),t,t) + 7*sp.diff(y(t),t) + 10*y(t) - sp.diff(x,t) - 6*x

initial_condition = {
    y(t).subs(t, 0): 6,
    y(t).diff(t).subs(t, 0): -4
}

sol = sp.dsolve(eq, ics=initial_condition)
sol
```

$$
y(t) = \left(\frac{44}{3} - 9e^{-t} + \frac{e^{-3t}}{3} \right)e^{-2t} = \frac{44}{3} e^{-2t} + \frac{1}{3} e^{-5t} - 9e^{-3t}
$$

Esse resultado coincide com o obitido no item (a), validando a solução analítica.
